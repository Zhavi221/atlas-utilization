```python
import uproot

# Sample file from CMS Open Data
sample_file = "root://eospublic.cern.ch//eos/opendata/cms/Run2016G/SingleElectron/MINIAOD/UL2016_MiniAODv2-v2/120000/003CE725-FA4B-FD47-B686-5889F0C27323.root"

print(f"Opening file: {sample_file}\n")

with uproot.open(sample_file) as file:
    # Get all trees
    all_trees = list(file.keys())
    
    for tree_name in all_trees:
        print("=" * 80)
        print(f"TREE: {tree_name}")
        print("=" * 80)
        
        tree = file[tree_name]
        branches = sorted(list(tree.keys()))
        
        print(f"\nTotal branches: {len(branches)}\n")
        
        for i, branch in enumerate(branches, 1):
            print(f"{i:4d}. {branch}")
        
        print("\n")
```
